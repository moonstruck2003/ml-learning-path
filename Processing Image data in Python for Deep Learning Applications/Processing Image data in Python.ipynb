```python
#resizing the image using Pillow 
from PIL import Image 
```

```python
img = Image.open('obl.jpeg')
img_resized = img.resize((200,200 ))
```

```python
img_resized.save('cat_image_resize.jpeg')
```

```python
img_plot = plt.imshow(img_resized)
plt.show()
```

```python
#Convert RGB images to greyscale image using OpenCV 
import cv2 
```

```python
img = cv2.imread('obl.jpeg')
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
cv2.imshow("Grayscale Image", grayscale_image)
cv2.waitKey(0)  # Keeps window open until a key is pressed
cv2.destroyAllWindows()  # Closes all OpenCV windows
```

```python
cv2.imwrite('grayscale_image.jpg',grayscale_image)
```
