```python
import pandas as pd 
import numpy as np 
import sklearn.datasets 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
```

```python
#loading the data 
breast_cancer_data = sklearn.datasets.load_breast_cancer()
print(breast_cancer_data)
```

```python
data_frame = pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label']=breast_cancer_data.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
data_frame['label'].value_counts()
```

```python
#0 Benign 
#1 Malignant 
#Splitting features and targets 
x = data_frame.drop(columns='label',axis=1)
y = data_frame['label']
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

```python
#GridSearchCV 
#GridSearchCV is used for determining the best parameters for our model 
#Load our SVC model 
model = SVC()
```

```python
#hyperparameters 
parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],  
    'C': [1,5,10,20] 
}
```

```python
#GridSearch 
classifier = GridSearchCV(model,parameters,cv=5)
```

```python
#Fitting the data to our model 
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
#best parameter for SVC 
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
#Determinin the higehst accuracy 
highest_accuracy = classifier.best_score_ 
print(highest_accuracy)
```

```python
# loading the results to pandas dataframe 
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
print(grid_search_result)
```

```python
#Highest Accuracy = 95.2% 
#Best Parameters = (C-10,kernal-linear)
```
