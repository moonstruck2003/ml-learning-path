```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#Label Enlcoding of breast cancer dataset

file_path= r"C:\Users\MRIDUL\Desktop\ML\data.csv"
cancer_data = pd.read_csv(file_path)
```

```python
cancer_data.head()
```

```python
#finding the count of M and B labels
cancer_data['diagnosis'].value_counts()
```

```python
#converting the labels as numerical values
#load the label encoder function
label_encode = LabelEncoder()
```

```python
#encoding all the labels as 0 or 1
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels ato the dataframe 
cancer_data['target']=labels
```

```python
cancer_data.head()
```

```python
# 0 represents B 
# 1 represents M
#value counting
cancer_data['target'].value_counts()
```
