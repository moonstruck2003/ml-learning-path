```python
#importing the dependencies
import numpy as np
import pandas as pd
file_path = r"C:\Users\MRIDUL\Desktop\ML\credit_data.csv"
credit_card_data = pd.read_csv(file_path)
credit_card_data.tail()
# 0 = legit 1 = fraud 
```

```python
#distribution of the two classes
credit_card_data['Class'].value_counts()
```

```python
#this is highly imbalanced dataset
# 0 = legit 1 = fraud 
```

```python
#seperating the 0 and 1
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(fraud.shape)
print(legit.shape)
```

```python
#Under-Sampling
# build a sample dataset containing similar distribution of legit and fraudlent 
#Number of fraud = 492

legit_sample = legit.sample(n=492)
legit_sample.shape
```

```python
# we need to concatanate the two dataframes frauds and legits
new_dataset = pd.concat([legit_sample,fraud], axis = 0)
new_dataset.head()
```
