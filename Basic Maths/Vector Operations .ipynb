```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
#Plotting vector
plt.quiver(0,0,4,5)
plt.show()
```

```python
#41+5j
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```

```python
#for mutliple vector
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,-6,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```

```python
#addition of two vectors, subtraction is also same
vector1= np.asarray([0,0,2,3])
vector2= np.asarray([0,0,3,-2])
sum = vector1+vector2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```

```python
#multiplying a vector by a scaler
vector1= np.asarray([0,0,2,3])
vector2 = 2*vector1
print(vector2)
```
