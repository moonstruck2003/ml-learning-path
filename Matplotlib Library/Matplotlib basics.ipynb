```python
#importing Matplotlib library
import matplotlib.pyplot as plt
#import numpy for data
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#plotting the data
#sine wave

plt.plot(x,y)
plt.show()
```

```python
#cosine wave

plt.plot(x,z)
plt.show()
```

```python
#adding titles x axis and y axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sin wave')
plt.show()
```

```python
#parabola curve
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
   #drawing multiple line
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g+')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

```python
#Bar PLot

fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
languages = ['English','Spanish','French','Latin','German']
people=[13,2451,213,5124,512]

ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('people')
plt.show()
```

```python
#pie chart

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','Spanish','French','Latin','German']
people=[13,2451,213,5124,512]

ax.pie(people, labels=languages, autopct= '%1.1f%%')
plt.show()
```

```python
#Scatter Plot

x=np.linspace(0,10,30)
y= np.sin(x)
z= np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='y')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D Scattered plot

fig3 = plt.figure()
ax=fig3.add_axes(111,projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
